## Método de Bisección

```python
def function(x):
  return 8*math.exp(1-x)+7*math.log(x) # Definimos nuestra función
```

```python
def MetodoBiseccion(a,b,nmax,ep,f=function): # Definimos la función de bisección
  """
  [a,b]: intervalo
  nmax: número máximo de iteraciones 
  ep: rango de incertidumbre
  """
  # Inicializamos el método de bisección
  i = 0 
  xa = (a+b)/(2)-ep
  xb = (a+b)/(2)+ep
  # Para colocar en un DataFrame (tablas) creamos listas con los resultados
  a_k = [a]
  b_k = [b]
  x_a = [xa]
  x_b = [xb]
  f_a = [f(xa)]
  f_b = [f(xb)]
  # comienza las iteraciones con condición de paro un número máximo de iteraciones
  while i <= nmax:
    if f(xa)<f(xb): # Dependiendo el condicional se lleva a repetir los métodos
      b = xb
      xa = (a+b)/(2)-ep
      xb = (a+b)/(2)+ep
      a_k.append(a)
      b_k.append(b)
      x_a.append(xa)
      x_b.append(xb)
    else:
      a = xa
      xa = (a+b)/(2)-ep
      xb = (a+b)/(2)+ep
      b_k.append(b)
      a_k.append(a)
      x_a.append(xa)
      x_b.append(xb)
    f_a.append(f(xa))
    f_b.append(f(xb))
    i += 1
  # Creamos un diccionario para posteriormente convertirlo en un DataFrame
  data = {
      "a":a_k,
      "b":b_k,
      "xa":x_a,
      "xb":x_b,
      "f(xa)":f_a,
      "f(xb)":f_b
  }
  df = pd.DataFrame.from_dict(data)
  return df
```

```python
# Colocamos los parametros indicados 
MetodoBiseccion(1,2,50,0.001) # [a,b,num_maximo_de_iteraciones,rango_de_incertidumbre]
```

```python
def function(x): # Declaramos la función
  return 8*math.exp(1-x)+7*math.log(x)

def primeraDerivada(x): #primera derivada
  return (7/x)-8*math.exp(1-x)

def segundaDerivada(x): #segunda derivada
  return (-7/x**2)+8*math.exp(1-x)
```

```python
def Newton(x0,TOL,f=function,dx=primeraDerivada,dx2=segundaDerivada):  
  """
  Para efectos prácticos
  x0 = x^{k}
  xk = x^{k+1}
  """
  # Iniciamos el método de Newton
  xk = x0 - (dx(x0))/(dx2(x0))
  # Creamos listas con los datos obtenidos para crear un DataFrame
  x_k = [xk]
  x_0 = [x0]
  abs_xk_x0 = [abs(xk-x0)] # error absoluto
  rel_xk_x0 = [abs((xk-x0)/x0)] # error relativo
  n = 0
  # Iniciamos el bucle con condición de paro el error relativo y un numero máximo de iteraciones n<100 en caso de no cumplir el error relativo
  while (abs((xk-x0)/x0)>TOL and n<100):
    n += 1
    # Cambiamos el valor de las variables y las guardamos en las listas
    x0 = xk 
    xk = x0 - (dx(x0))/(dx2(x0))
    relativo = (xk-x0)/x0
    x_k.append(xk)
    x_0.append(x0)
    abs_xk_x0.append(abs(xk-x0))
    rel_xk_x0.append(abs((xk-x0)/x0))
  # Creamos un diccionario con las listas para posteriormente convertirlo en un DataFrame
  data2 = {
      "xk":x_0,
      "xk+1":x_k,
      "(xk_1-xk)":abs_xk_x0, # error absoluto
      "(xk_1-xk)/xk":rel_xk_x0 # error relativo
  }
  df2 = pd.DataFrame.from_dict(data2)
  return df2
```

##Método Secante

```python
def function(x): # Definimos la función
  return 8*math.exp(1-x)+7*math.log(x)

def primeraDerivada(x): #primera derivada
  return (7/x)-8*math.exp(1-x)
```

```python
def secante(x0,x1,TOL,f=function,dx=primeraDerivada):
  """
  Para efectos prácticos
  x0 = x^{k-1}
  x1 = x^{k}
  xk = x^{k+1}
  """
  # Iniciamos el método de Secante con su formula
  k = 0
  xk = (x0*dx(x1)- x1*dx(x0))/( dx(x1)-dx(x0))
  # Guardamos los resultados en listas
  x_k = [xk]
  x_0 = [x0]
  x_1 = [x1]
  f_x0 = [dx(x0)] #f'(xk-1)
  f_x1 = [dx(x1)] # f'(xk)
  abs_xk_x1 = [abs(xk-x1)] # error absoluto
  rel_xk_x1 = [abs((xk-x1)/x1)] # error relativo
  # Iniciamos el bucle while como condición de paro el error relativo y un numero maximo de iteraciones k<100
  while (abs((xk-x1)/x1)>=TOL and k<100):
    # Se les modifican los valores para continuar con la iteración y se guardan en las listas los resultados obtenidos
    k +=1 
    x0 = x1
    x1 = xk
    xk = (x0*dx(x1)- x1*dx(x0))/( dx(x1)-dx(x0))
    x_k.append(xk)
    x_0.append(x0)
    x_1.append(x1)
    f_x0.append(dx(x0))
    f_x1.append(dx(x1))
    abs_xk_x1.append(abs(xk-x1))
    rel_xk_x1.append(abs((xk-x1)/x1))
  # Creamos un diccionario con las listas para hacer despues un DataFrame
  data3 = {
    "xk-1":x_0,
    "xk":x_1,
    "xk+1":x_k,
    "f'(xk-1)":f_x0,
    "f'(xk)":f_x1,
    "(xk_1-xk)":abs_xk_x1,
    "(xk_1-xk)/xk":rel_xk_x1
  }
  df3 = pd.DataFrame.from_dict(data3)
  return df3
```

```python
# Colocamos los parametros indicados
secante(1,2,1e-6) #[x0,x1,tolerancia]
```
